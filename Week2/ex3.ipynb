```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
```

Example 3-1

```python
lines_1 = sc.textFile("../Data/ignatian_pedagogy")
```

```python
words_1 = lines_1.flatMap(lambda line : line.split())
```

```python
distinct_words_1 = words_1.distinct()
```

Exmple 3-2

```python
lines_2 = sc.textFile("../Data/README.md")
```

```python
words_2 = lines_2.flatMap(lambda line : line.split())
```

```python
distinct_words_2 = words_2.distinct()
```

```python
distinct_words_1.union(distinct_words_2)
```

```python
distinct_words_1.intersection(distinct_words_2)
```

```python
distinct_words_1.subtract(distinct_words_2).collect()
```

```python
distinct_words_1.cartesian(distinct_words_2).collect()
```

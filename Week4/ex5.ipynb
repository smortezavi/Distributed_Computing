```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
```

```python
lines = sc.textFile("../Data/filtered_registered_business_sf.csv")
```

```python
pairs = lines.map(lambda x: (x.split(",")[0], (x.split(",")[1],x.split(",")[3])))
```

```python
pairs.collect()
```

```python
pairs.countByKey()
```

```python
pairs.collectAsMap()
```

```python
pairs.lookup('')
```

```python
len(pairs.lookup(''))
```

```python
pairs.filter(lambda x: "San Francisco" not in x[1][1]).collect()
```
